```python
balance = 1000

while balance > 0:
    print(f"\n💰 Current balance: ₹{balance}")
    amount = int(input("Enter withdrawal amount (0 to quit): "))

    if amount == 0:
        print("Thank you! Goodbye.")
        break    # We'll learn about break soon!
    elif amount > balance:
        print("❌ Insufficient balance!")
    else:
        balance = balance - amount
        print(f"✅ ₹{amount} withdrawn successfully.")

if balance == 0:
    print("⚠️ Your balance is now ₹0.")
```
